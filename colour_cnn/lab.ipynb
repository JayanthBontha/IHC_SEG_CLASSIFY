```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('finaldata_lab.csv')
```

```python
blue_rows = df[df["Prediction"] == "Blue"]
non_blue_rows = df[df["Prediction"] == "Brown"]
```

```python
def plotter(blue_rows,non_blue_rows,i1,i2):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Plotting the line graph for blue rows
    axes[0].plot(blue_rows[i1], blue_rows[i2], 'o', label='Blue Prediction')
    axes[0].plot(blue_rows[i2], blue_rows[i2], label='45 Degree Line')
    axes[0].set_xlabel(i1)
    axes[0].set_ylabel(i2)
    axes[0].legend()

    # Plotting the line graph for non-blue rows
    axes[1].plot(non_blue_rows[i1], non_blue_rows[i2], 'o', color='red', label='Non-Blue Prediction')
    axes[1].plot(non_blue_rows[i2], non_blue_rows[i2], color='red', label='45 Degree Line')
    axes[1].set_xlabel(i1)
    axes[1].set_ylabel(i2)
    axes[1].legend()

    # Displaying the plot
    plt.tight_layout()
    plt.show()
```

```python
plotter(blue_rows,non_blue_rows,"L","B")
```

```python
filtered_df = df[(df['L'] > 20) &
                 (df['L'] <= 80) &
                 (df['A'] > 0) &
                 (df['A'] < 17) &
                 (df['B'] <= 0)]

# Display the filtered DataFrame
print(filtered_df)
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame based on the color prediction
blue_df = df[df['Prediction'] == 'Blue']
brown_df = df[(df['Prediction'] == 'Brown') & (df['FinalPred']=="W")]

# Create a graph for blue predictions
plt.figure(figsize=(10, 6))
plt.scatter(blue_df['A'] + blue_df['B'], blue_df['L'], color='blue')
plt.title('Blue Predictions with A+B')
plt.xlabel('A + B')
plt.ylabel('L')
plt.show()

# Create a graph for brown predictions
plt.figure(figsize=(10, 6))
plt.scatter(brown_df['A'] + brown_df['B'], brown_df['L'], color='brown')
plt.title('Brown Predictions with A+B')
plt.xlabel('A + B')
plt.ylabel('L')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame based on the color prediction
blue_df = df[df['Prediction'] == 'Blue']
brown_df = df[(df['Prediction'] == 'Brown') & (df['FinalPred'] == "W")]

# Create a graph with points of different colors
plt.figure(figsize=(10, 6))

# Plot blue predictions
plt.scatter(blue_df['A'] / blue_df['B'], blue_df['L'], color='blue', label='Blue Predictions')

# Plot brown predictions
plt.scatter(brown_df['A'] / brown_df['B'], brown_df['L'], color='brown', label='Brown Predictions')

plt.title('Predictions with A+B')
plt.xlabel('A + B')
plt.ylabel('L')
plt.legend()
plt.show()
```
